#### Pandas-DataFrame And Series
Pandas is a powerful data manipulation library in Python, widely used for data analysis and data cleaning. It provides two primary data structures: Series and DataFrame. A Series is a one-dimensional array-like object, while a DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns).

```python
import pandas as pd
import numpy as np

# Verify installation
print(pd.__version__)  
```

```python
import pandas as pd

# Create a simple list
data = [1, 2, 3, 4, 5]

# Convert list to Series
series = pd.Series(data)

# Display the Series
print(series)
```

```python
## Series
## Pandas Series is a one-dimensional array-like object that can hold any data type. It is similar to a column in a table.
print(type(series))
```

```python
# Create a dictionary
data = {'a': 1, 'b': 2, 'c': 3}

# Dictionary keys become the INDEX
series_dict = pd.Series(data)
print(series_dict)
```

```python
data = [10, 20, 30]
my_index = ['a', 'b', 'c']

# Provide custom index explicitly
series_custom = pd.Series(data, index=my_index)
print(series_custom)    
```

```python
# Integer Series
int_series   = pd.Series([1, 2, 3, 4])   
print(int_series)    
```

```python
# Float Series
flt_series   = pd.Series([1.1, 2.2, 3.3])      
print(flt_series)
```

```python
# String Series
str_series   = pd.Series(['Asbin', 'Binod'])     
print(str_series)
```

```python
# Boolean Series
bool_series  = pd.Series([True, False, True])  
print(bool_series)
```

```python
mix_series   = pd.Series([1, 'two', 3.0]) 
print(mix_series)
```

```python
s = pd.Series([10, 20, 30, 40], index=['a','b','c','d'])

print(s.dtype)      
print(s.shape)      
print(s.size)    
print(s.index)    
print(s.values)  
print(s.is_unique)
```

```python
s1 = pd.Series([10, 20, 30])
s2 = pd.Series([1, 2, 3])

# Arithmetic — element-wise (vectorized)
print(s1 + s2)  
print(s1 - s2)  
print(s1 * s2) 
print(s1 / s2)  
```

```python
# Scalar operations
print(s1 + 100) 
print(s1 ** 2)    

# Boolean / comparison
print(s1 > 15)    
```

```python
s = pd.Series([10, 20, 30, 40, 50], index=['a','b','c','d','e'])

# By label (loc) 
print(s['a'])         
print(s[['a', 'c']])   
print(s['a':'c'])     
```

```python
# By position (iloc)
print(s[0])      
print(s[0:3])        
print(s[-1])         
```

```python
# Boolean indexing
print(s[s > 25])
```

```python
df = pd.DataFrame({'Name':['Krishna','John','Jack','Sara'],
                   'Age':[25,30,45,22], 'Salary':[50000,60000,90000,35000]})

# Single condition 
df[df['Age'] > 28]                   
```

```python

df[(df['Age'] > 25) & (df['Salary'] > 50000)]   
```

```python
# OR condition
df[(df['Age'] < 26) | (df['Age'] > 40)]         
```

```python
# NOT condition 
df[~(df['Name'] == 'Jack')]          
```

```python
# .isin() filter
df[df['Name'].isin(['Krish', 'Sara'])]
```

```python
# .query() — more readable syntax 
df.query('Age > 28 and Salary > 50000')
```
